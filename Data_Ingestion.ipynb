#Data Visualization
Plot findings

```python
# Create a boxplot to visualize distribution of `video_duration_sec`
ax=data.boxplot(column="video_duration_sec")
ax.set_ylabel('Duration')
```

```python
# Create a histogram f the values in the video_duration_sec column to further explore the distribution of this variable.
du = data['video_duration_sec']
plt.hist(du, bins=12)
plt.show()
```
